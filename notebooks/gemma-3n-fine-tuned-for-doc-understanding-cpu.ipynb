# CPU Notebook - FLAN-T5 for Invoice Doc Understanding
Version CPU-compatible locale basee sur `data/kaggle_invoice_images` (OCR text -> JSON).


````python
from pathlib import Path
DATA_ROOT = Path('data/kaggle_invoice_images').resolve()
assert DATA_ROOT.exists(), f'Data root introuvable: {DATA_ROOT}'
print('DATA_ROOT =', DATA_ROOT)
````

````python
import csv
csvs = list(DATA_ROOT.rglob('*.csv'))
print('CSV count =', len(csvs))
for p in sorted(csvs)[:5]:
    print('-', p)
````

````python
from src.doc_understanding_cpu.config import CPUTrainConfig
from src.doc_understanding_cpu.train import run_cpu_training

cfg = CPUTrainConfig(
    data_root=DATA_ROOT,
    output_dir=Path('artifacts/doc_understanding_cpu').resolve(),
    model_name='google/flan-t5-small',
    max_train_samples=128,
    max_eval_samples=64,
    num_train_epochs=25,
    per_device_train_batch_size=2,
    per_device_eval_batch_size=2,
)
summary = run_cpu_training(cfg)
summary
````

````python
from src.doc_understanding_cpu.config import CPUInferConfig
from src.doc_understanding_cpu.inference import load_cpu_predictor

predictor = load_cpu_predictor(CPUInferConfig(artifacts_dir=Path('artifacts/doc_understanding_cpu').resolve()))
ocr_text = 'Invoice Number: INV-2026-001\nInvoice Date: 2026-01-12\nClient: ACME Corp'
pred = predictor.predict(ocr_text)
pred
````

````python
from src.doc_understanding_cpu.data import load_cpu_records
from src.doc_understanding_cpu.eval import evaluate_cpu_predictions

eval_records = load_cpu_records(DATA_ROOT / 'batch_1/batch_1/batch1_2.csv')[:10]
pred_map = {}
for r in eval_records:
    out = predictor.predict(r.ocr_text)
    pred_map[r.img_name] = out['prediction'] if isinstance(out.get('prediction'), dict) else {}
metrics = evaluate_cpu_predictions(eval_records, pred_map)
metrics
````
